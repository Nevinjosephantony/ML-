```python
import pandas as pd
import matplotlib.pyplot as plt

# Read dataset
df = pd.read_csv("Student_Performance.csv")

# Features and labels
x = df["Hours Studied"].values
y = df["Performance Index"].values

# Initialize parameters
m = 0
b = 0

# Hyperparameters
learning_rate = 0.0001
epochs = 1000

# Number of data points
n = len(x)

# Training loop
for i in range(epochs):

    m_gradient = 0
    b_gradient = 0

    for j in range(n):

        # Prediction
        y_pred = m * x[j] + b

        # Error
        error = y[j] - y_pred

        # Gradients
        m_gradient += -(2/n) * x[j] * error
        b_gradient += -(2/n) * error

    # Update parameters
    m = m - learning_rate * m_gradient
    b = b - learning_rate * b_gradient

# Final equation
print("Slope:", m)
print("Intercept:", b)

# Prediction
hours = 6
prediction = m * hours + b

print("Predicted Score:", prediction)

# Plot graph
plt.scatter(x, y)

y_line = m * x + b

plt.plot(x, y_line, color='red')

plt.xlabel("Hours Studied")
plt.ylabel("Performance Index")

plt.show()
```
